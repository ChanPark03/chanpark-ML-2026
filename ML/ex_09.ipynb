```python
import numpy as np
```

```python
X = np.array([['Korea', 44, 7200], ['Japan', 27, 4800], ['China', 30, 6100]])
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
```

```python
# мӣҗн•ҳлҠ” м—ҙмқ„ лҪ‘м•„м„ң 2м°Ёмӣҗ л°°м—ҙлЎң л§Ңл“Өм–ҙм„ң м „лӢ¬н•ҳм—¬м•ј н•ңлӢӨ.
XX = onehotencoder.fit_transform(X[:,0].reshape(-1,1)).toarray()
print(XX)
```

```python
class_vector = [2, 6, 6, 1]
```

```python
X = np.delete(X, [0], axis=1) # 0лІҲм§ё м—ҙ мӮӯм ң
X = np.concatenate((XX, X), axis = 1) # XмҷҖ XXлҘј л¶ҷмқёлӢӨ.
print(X)
```

```python
import keras
mnist = keras.datasets.mnist
```

```python
(X_train, y_train),(X_test, y_test) = mnist.load_data()
(X_train, X_test) = (X_train / 255.0, X_test / 255.0)
```

```python
model = keras.Sequential()
model.add(keras.layers.Input(shape=(28, 28)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(512, activation='relu'))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(10, activation='softmax'))#Sparse Categorial Crossentropy
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20)
model.evaluate(X_test, y_test)
```
