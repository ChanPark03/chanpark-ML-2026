```python
import keras
```

```python
mlp = keras.models.Sequential()
```

```python
input_layer = keras.layers.Input(shape = (2,))
dense1  = keras.layers.Dense(32, activation = 'sigmoid')
dense2 = keras.layers.Dense(1, activation = 'sigmoid')
mlp.add(input_layer)
mlp.add(dense1)
mlp.add(dense2)
mlp.summary()
```

```python
X = tf.constant([[0,0], [0,1], [1,0], [1,1]])
y = tf.constant([0, 1 ,1 ,0])

```

```python
mlp.compile(loss='mse', optimizer='sgd')
mlp.fit(X, y, epochs= 1000)
```

```python
print(mlp.predict(X))
```
