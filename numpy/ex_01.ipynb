```python
def train_test_split(numbers):
    np.random.shuffle(numbers)
    divisor = int(len(numbers) * 0.8)
    train = numbers[:divisor]
    test = numbers[divisor:]
    return (train, test)
x = np.arange(1,51)
(train, test) = train_test_split(x)
print(f'train: {train}, \t Test: {test}')
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, 'r-')
plt.plot(x, y2, 'b:')
plt.savefig("Sine-Cosine.pdf")
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, 'r-', label='sin(x)')
plt.plot(x, y2, 'b:', label='cos(x)')
plt.legend()
plt.show()
```
